```python
def recursive_json(json_data, list_data, columns, ticker_list):

    for i in json_data.keys():
        if(not(isinstance(json_data[i], dict))):
            if(i == 'symbol' and json_data[i] not in ticker_list):
                return None, None
            columns.append(i)
            list_data.append(json_data[i])
        else:
            for j in json_data[i].keys():
                for k in range(len(json_data[i][j])):
                    recursive_json(json_data[i][j][k], list_data, columns, ticker_list)
    
    return list_data, columns
```
